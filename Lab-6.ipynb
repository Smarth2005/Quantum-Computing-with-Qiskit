```python
# Import libraries
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_vector
import numpy as np
import matplotlib.pyplot as plt
```

```python
def blochCoordinates(state):
  a, b = state.data

  x = 2 * np.real(np.conj(a) * b)
  y = 2 * np.imag(np.conj(b) * a)
  z = np.abs(a)**2 - np.abs(b)**2

  bloch_vector = [x, y, z]
  return bloch_vector
```

### Experiment-1: X-Gate on Bloch Sphere and State Vector

Create a quantum circuit with single qubit using the Qiskit framework and perform the following tasks:

2. Draw the Circuit.
3. Show the qubit on Bloch Sphere.

### $X|0\rangle = |1\rangle$

```python
qc = QuantumCircuit(1)
qc.x(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $X|0\rangle = |1\rangle$", font_size=12)
```

### Experiment 2: Single Qubit Gates on Bloch Sphere and State Vector
Repeat the experiment 1 for Y, Z, Phase gate, S, S†, T, T† gates with initial state vector
$$
|\psi_i\rangle = \frac{1}{\sqrt{2}} (|0\rangle + |1\rangle)
$$

We can take initial state as $\boxed{|\psi_i\rangle = H|0\rangle}$

### $(i)\, Y|\psi_i\rangle$

$Y|\psi_i\rangle = \frac{1}{\sqrt2}(Y|0\rangle + Y|1\rangle)$

$Y|\psi_i\rangle = \frac{1}{\sqrt2}(i|1\rangle -i|0\rangle)$

${Y|\psi_i\rangle = \frac{i}{\sqrt2}(|1\rangle -|0\rangle)}$

```python
qc = QuantumCircuit(1)

# Create superposition
qc.h(0)

# Apply gate
qc.y(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $Y|\psi_i\rangle$", font_size=12)
```

### $(ii)\, Z|\psi_i\rangle$

$Z|\psi_i\rangle = \frac{1}{\sqrt2}(Z|0\rangle + Z|1\rangle)$

$Z|\psi_i\rangle = \frac{1}{\sqrt2}(|0\rangle - |1\rangle)$

```python
qc = QuantumCircuit(1)

# Create superposition (initial state)
qc.h(0)

# Apply gate
qc.z(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $Z|\psi_i\rangle$", font_size=12)
```

### $(iii)\, P(\phi)|\psi_i\rangle$

We take $\boxed{\phi = \frac{\pi}{2}}$ ~ S gate.

```python
qc = QuantumCircuit(1)

qc.h(0)
qc.p(np.pi/2, 0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $P(\phi)|\psi_i\rangle$", font_size=12)
```

### $(iv)\, S|\psi_i\rangle$

$S|\psi_i\rangle = \frac{1}{\sqrt2}(S|0\rangle + S|1\rangle)$

$S|\psi_i\rangle = \frac{1}{\sqrt2}(|0\rangle + i|1\rangle)$

```python
qc = QuantumCircuit(1)

qc.h(0)
qc.s(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $S|\psi_i\rangle$", font_size=12)
```

### $(v)\, S^\dagger|\psi_i\rangle$

$S^\dagger|\psi_i\rangle = \frac{1}{\sqrt2}(S^\dagger|0\rangle + S^\dagger|1\rangle)$

$S^\dagger|\psi_i\rangle = \frac{1}{\sqrt2}(|0\rangle -i|1\rangle)$

```python
qc = QuantumCircuit(1)

qc.h(0)
qc.sdg(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $S^\dagger|\psi_i\rangle$", font_size=12)
```

### $(vi)\, T|\psi_i\rangle$

$T|\psi_i\rangle = \frac{1}{\sqrt2}(T|0\rangle + T|1\rangle)$

$T|\psi_i\rangle = \frac{1}{\sqrt2}(|0\rangle + e^{i\pi/4}|1\rangle)$

```python
qc = QuantumCircuit(1)

qc.h(0)
qc.t(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $T|\psi_i\rangle$", font_size=12)
```

### $(vii)\, T^\dagger|\psi_i\rangle$

$T^\dagger|\psi_i\rangle = \frac{1}{\sqrt2}(T^\dagger|0\rangle + T^\dagger|1\rangle)$

$T^\dagger|\psi_i\rangle = \frac{1}{\sqrt2}(|0\rangle - e^{i\pi/4}|1\rangle)$

```python
qc = QuantumCircuit(1)

qc.h(0)
qc.tdg(0)

print("Quantum Circuit:")
display(qc.draw('mpl'))

# Get statevector
state = Statevector.from_instruction(qc)
print("Statevector:")
print(state)

# Convert statevector to Bloch coordinates and plot bloch sphere
bloch_vector = blochCoordinates(state)

plot_bloch_vector(bloch_vector, title=r"Bloch Sphere: $T^\dagger|\psi_i\rangle$", font_size=12)
```

### Circuit (a)

```python
from qiskit import QuantumCircuit
from qiskit.primitives import StatevectorSampler
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

qc = QuantumCircuit(2)

qc.h(1)
qc.cz(0, 1)
qc.h(1)

qc.measure_all()
display(qc.draw('mpl'))

# Execute using the Statevector Sampler
sampler = StatevectorSampler()
job = sampler.run([qc], shots=1024)
result = job.result()

# Extract and Plot Results
# In V2 primitives, data is accessed via result[index].data
counts = result[0].data.meas.get_counts()
print("Measurement Counts:", counts)

plot_histogram(counts)
```

```python
# Required Imports
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit_aer import AerSimulator
from qiskit_ibm_runtime import QiskitRuntimeService, EstimatorV2 as Estimator
```

### Simulation using AerSimulator

```python
# Using local simulator for initial testing
backend = AerSimulator()
print(f"Using backend: {backend}")

qc = QuantumCircuit(2)
qc.h(1)
qc.cz(0, 1)
qc.h(1)

display(qc.draw('mpl'))

# Define the observable
# We measure the expectation value of ZZ (Z on both qubits)
observable = SparsePauliOp("ZZ")

# Execution using simulator
estimator = Estimator(mode=backend)

# We pass a list of (circuit, observable) pairs
pub = (qc, observable)
job = estimator.run([pub])
result = job.result()[0]

# Results
print(f"Expectation Value <ZZ> for Circuit (a): {result.data.evs}")
```

### Execution on Real Hardware

```python
import os
from dotenv import load_dotenv

# Load the variables from .env into the environment
load_dotenv()

# Fetch the token using the key name
api_token = os.getenv("IBM_TOKEN")
crn_instance = os.getenv("CRN")
```

```python
from qiskit_ibm_runtime import QiskitRuntimeService

QiskitRuntimeService.save_account(
token = api_token, # Use the 44-character API_KEY you created and saved from the IBM Quantum Platform Home dashboard
instance = crn_instance,
)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import SparsePauliOp
from qiskit.transpiler.preset_passmanagers import generate_preset_pass_manager
from qiskit_ibm_runtime import QiskitRuntimeService, EstimatorV2 as Estimator

# Initialize service and backend
service = QiskitRuntimeService()
backend = service.least_busy(simulator=False, operational=True)

# ISA Circuit & Pass Manager
qc = QuantumCircuit(2)
qc.h(1)
qc.cz(0, 1)
qc.h(1)

print("Quantum circuit:\n")
display(qc.draw("mpl"))

pm = generate_preset_pass_manager(backend=backend, optimization_level=1)
isa_circuit = pm.run(qc)

print("ISA circuit:\n")
display(isa_circuit.draw("mpl", idle_wires=False))

# ISA Observable
observable = SparsePauliOp("ZZ")
isa_observable = observable.apply_layout(isa_circuit.layout)

# Run Estimator
estimator = Estimator(mode=backend)
estimator.options.resilience_level = 1 # Error mitigation
estimator.options.default_shots = 5000

# We submit one PUB: (the circuit, the observable)
job = estimator.run([(isa_circuit, isa_observable)])

print(f">>> Job ID: {job.job_id()}")
result = job.result()[0]

print(f"Expectation Value <ZZ> for circuit(a): {result.data.evs}")
```
